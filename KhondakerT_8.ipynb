```python
import pandas as pd
import seaborn as sns
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.head()
```

```python
avocado.info(memory_usage='deep')
```

```python
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.info(memory_usage='deep')
```

```python
av1 = avocado.query('year == 2015 and region == "Albany" and type == "conventional"')[['Date','Total Bags','Small Bags']]
av1.head()
```

```python
avocado_melted = pd.melt(avocado, id_vars='Date', value_vars=['Total Bags', 'Small Bags'], var_name='Bags', value_name='Count')
avocado_melted
```

```python
sns.relplot(data=avocado_melted, kind='line', x='Date', y='Count', hue='Bags')
```

```python
avocado['Sales Volume'] = pd.qcut(avocado['Total Volume'], q=4, labels=['Poor','Modest','Good','Excellent'])
avocado.head()
```

```python
sns.countplot(data=avocado, x='year', hue='Sales Volume')
```

```python
avocado_melted_bag = pd.melt(avocado, id_vars='year', value_vars=['Small Bags','Large Bags','XLarge Bags'], var_name='Bags', value_name='Count')
avocado_melted_bag
```

```python
sns.relplot(data=avocado_melted_bag, kind='line', x='year', y='Count', hue='Bags')
```
